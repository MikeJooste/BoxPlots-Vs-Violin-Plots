```python
#Import Pandas for dataframe processing, Os for workind directory changes, and seaborn for visualisations
import pandas as pd 
import os 
import seaborn as sns
from matplotlib import pyplot as plt 
%matplotlib inline 
import warnings 
warnings.filterwarnings('ignore')
os.chdir('C:\\Users\\mike\\documents\Python Training')
os.getcwd()
movies= pd.read_csv('P4-Movie-Ratings.csv')
movies
#rename columns
movies.columns=['Film', 'Genre','CriticRating','AudienceRating', 'Budget', 'Year']
movies.head()
Filter1=movies.Genre=='Action'
Filter2=movies.CriticRating>30
movies[Filter1&Filter2]
movies.head()
```

```python
#Violin Plots 
z=sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year',y='CriticRating').set_title('Critic Rating Range by Year_ViolinPlot')
```

```python
#Violin Plots 2
w=sns.boxplot(data=movies[movies.Genre=='Drama'], x='Year',y='CriticRating').set_title('Critic Rating Range by Year_BoxPlot')
```

```python
#Subplots are very useful 
f, axes =plt.subplots(1,2,figsize=(15,15),sharex=True)
sns.set_style(('white'), {'axes.facecolor':"gray"})
z=sns.violinplot(data=movies[movies.Genre=='Drama'], x='Year',y='CriticRating',ax=axes[0]).set_title('Critic Rating Range by Year_ViolinPlot')
w=sns.boxplot(data=movies[movies.Genre=='Drama'], x='Year',y='CriticRating',ax=axes[1]).set_title('Critic Rating Range by Year_BoxPlot')
plt.show()
```

```python
#Finishing Touches Thematic edits 
import seaborn as sns
from matplotlib import pyplot as plt
movies['Genre']=movies.Genre.astype('category')
list1 =list()
mylabels=list()
for gen in movies.Genre.cat.categories: 
    list1.append(movies[movies.Genre==gen].Budget)
    mylabels.append(gen)
sns.set_style('whitegrid')
fig, ax =plt.subplots()
fig.set_size_inches(11.7, 8.27) #size of A4 paper 
   
h=plt.hist(list1,bins=30, stacked=True,rwidth=1, label=mylabels)
plt.title('Movie Budget Distribution', fontsize=30,\
         color='black', fontname='Console')
plt.ylabel('Numbe of Movies', fontsize=20,color='red')
plt.xlabel("Budget", fontsize=20, color='green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fancybox= True,shadow=True, prop={'size':20})
plt.show()
```
